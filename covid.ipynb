# COVID-19 Data Filtering and Ploting by Contry 

- Linear regression using below equations
- Ploting HiLo plots where real value will move around regression line

```j
NB. Required library scripts
require 'tables/csv'
require 'web/gethttp'
require 'graphics/plot'
load 'math/calculus'

CANVAS_DEFSIZE_jwplot_ =: 700 350

NB. Utility verbs
chop =: 4 : '(y#i.#y)</.x'
NB. desc =: 3 : '>(I. (<y) E. 0{keys) { 1{keys'
get =: 4 : 0
  'a b' =: y
  (I. (<b) E. a{|:x){x 
)
gets =: 4 : '(I. (<y) E. 0{|:x){x'
gety =: 4 : '(I. (<y) E. 1{|:x){x'
getm =: 4 : '(I. (<y) E. 4{|:x){x'
vfmx =: 3 : '}:(,|."1 [ 1,.-. *./\"1 |."1 y='' '')#,y,.LF'
lreg =: 4 : 'y %. 1 ,. x'
treg =: 3 : '((1{((i.#y) lreg y))*(i.#y))+(0{((i.#y) lreg y))'


col =: dyad : 'makenum }. x {"1 y'
columns =: monad : '(;/ |: ,: i. #0{y) ,. (|: ,: 0{ y)'
counter =: monad : '(~. y) ,. (;/ (#/.~ y))'
```

```j
covid =: fixcsv gethttp dquote 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'
```

```j
ma =: 3 : '10 (+/%#)\ 5 col covid  gets y' NB. moving average of 10 days of daily cases

plot (ma 'EGY'),:(ma 'OMN'),: (ma 'KWT')
```

```j
cov=: 70}. 5 col covid  gets 'EGY'
covv =: cov gets 0
reg =: treg cov
ann =: {: cov
pd 'reset'
pd 'color orange; type hilo; pensize 7' 
pd 'text -200x 310x Last Day=',(":ann)
pd cov,:reg 
pd 'show'
```
